## Session 2 - Simple text processing

```python
colours =  ["red", "green", "blue"]
# list = [a1, a2, a3]
```

### for loop

```python
for colour in colours:
    print(colour)
```

```python
for colour in colours:
    print(colour.upper() + "!!")
```

```python
colours_shouted = []

for colour in colours:
    colour_upper = colour.upper() # make upper case
    colour_shouted = colour_upper + "!!" # make exclamaition points
    colours_shouted.append(colour_shouted) # add to the new, empty list
    print(colour_shouted)
```

### if/else

```python
for colour in colours:
    if colour == "red":
        print("*" + colour + "*")

    elif colour == "green":
        print(colour.upper() + "!!")

    else:
        print(colour)
```

```python
#  The list consists of different types - exception handling:

weird_list = ["red", "green", 10, "blue"]

for value in weird_list:
    if type(value) == int:
        print(value/2)
    else:
        pass # makes it skip the lines, that isnt int
        #print("not a number!")
```

### Exception handling

```python
for value in weird_list:
    try:
        print(value/2)
    except:
        print("not a number!")
```
